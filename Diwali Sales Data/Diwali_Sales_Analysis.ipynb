# Exploratory Data Analysis

### Gender

```python
df.columns
```

```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(x = 'Gender',data = df)

for bars in ax.containers:
    ax.bar_label(bars)



# sns = seaborn(a Python visualization library), to make a bar chart.(When Seaborn draws bars, each bar is stored inside something called a "container".)
# countplot = bar chart of counts(Creates a bar chart showing counts of categories like "Male", "Female")
#x='Gender' → Uses the 'Gender' column from your DataFrame for the x-axis (Male/Female).It counts how many times each category (like "Male", "Female") appears.
#data=df → Tells Seaborn to use your DataFrame df as the source.
# ax = → varible to Stores the chart object/plot object (axes)
#container → Each bar that Seaborn draws gets stored inside a "container."
# ax.containers = is just a collection (list) of those bars, which you can loop through to add labels (holds bar objects)
# bar_label = show count labels on bars(like F = 7832 & m = 3407)    
```

```python
ax = sns.countplot(x='Gender', data=df, palette={'F':'pink','M':'lightblue'})

# palette = defines color scheme for plot
# {'F':'pink','M':'lightblue'} = dictionary of key:value pairs
```

```python
ax = sns.countplot(x='Gender', data=df, palette='Set2')

#palette →  set of colors.(color scheme for your chart)
#Seaborn has many palettes like 'Set1', 'Set2', 'pastel', 'muted', 'coolwarm'.

```

```python
ax = sns.countplot(x='Gender', data=df, palette='Set1')
```

```python
ax = sns.countplot(x='Gender', data=df, palette='coolwarm')
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# groupby('Gender') = split data by Gender
# as_index=False = keep Gender as column see ex. below

# Example with as_index=True (default):
# Result:
#          Amount
# Gender          
# F          7832
# M          3407
# Here, Gender is used as the index (row labels).

# Example with as_index=False:
# Result:
#      Gender  Amount
# 0      F    7832
# 1      M    3407
# Here, Gender stays as a normal column.

# ['Amount'] = pick Amount column
# sum() = add values for each Gender
# sort_values(by='Amount') = order by Amount
# ascending=False = largest first
```

```python
# plotting a bar chart for gender vs total amount

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen, palette={'F':'pink','M':'lightblue'})



# sales_gen = create new DataFrame
# df.groupby(['Gender'], as_index=False) = group data by Gender, keep Gender as column
# ['Amount'].sum() = add up Amount for each Gender
# .sort_values(by='Amount', ascending=False) = sort totals by Amount, largest first

# sns.barplot(...) = draw bar chart
# x='Gender' = categories on x-axis
# y='Amount' = totals on y-axis
# data=sales_gen = use grouped/sorted data
# palette={'F':'pink','M':'lightblue'} = set colors: F=pink, M=lightblue
```

*From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men*

### State

```python
# total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders',palette='Set2')


# sales_state = create new DataFrame
# df.groupby(['State'], as_index=False) = group data by State, keep State as column
# ['Orders'].sum() = add up Orders for each State
# .sort_values(by='Orders', ascending=False) = sort totals by Orders, largest first
# .head(10) = keep only top 10 States

# sns.set(rc={'figure.figsize':(15,5)}) = set chart size to 15 width x 5 height
      # [sns.set(rc={'figure.figsize':(15,5)}) = change default chart size
      # figure.figsize = controls width and height of the plot
      # (15,5) = width=15 units, height=5 units
      # makes chart wider so long State names fit better on x-axis]
      # rc = runtime configuration dictionary (fixed keyword)


# sns.barplot(...) = draw bar chart
# data=sales_state = use grouped/sorted data
# x='State' = show State names on x-axis
# y='Orders' = show total Orders on y-axis

```

```python
# total amount/sales from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount',palette='Set2')
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*
